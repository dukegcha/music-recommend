Text Cleaning / Text Preprocessing
Remove \n, and replace all characters to lower case

```python
df['text'].str.lower().replace(r'^\w\s', '').replace(r'\n',' ', regex = True)
```

```python
df['text'] = df['text'].str.lower().replace(r'^\w\s', '').replace(r'\n',' ', regex = True)
```

```python
df
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
def token(txt):
    token = nltk.word_tokenize(txt)
    a = [stemmer.stem(w) for w in token]
    return " ".join(a)
```

```python
token("you are beautiful, beauty")
```

```python
df['text'].apply(lambda x: token(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfid = TfidfVectorizer(analyzer='word', stop_words='english')
```

```python
matrix = tfid.fit_transform(df['text'])
```

```python
similar = cosine_similarity(matrix)
```

```python
similar[0]
```

Recommender Function

```python
def recommendation(song_df):
    idx = df[df['song'] == song_df].index[0]
    distances = sorted(list(enumerate(similar[idx])),reverse=True,key=lambda x:x[1])
    
    songs = []
    for m_id in distances[1:21]:
        songs.append(df.iloc[m_id[0]].song)
        
    return songs
```

```python
recommendation('Panda Bear')
```
